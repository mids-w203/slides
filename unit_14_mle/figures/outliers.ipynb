```r
library(ggplot2)
library(magrittr)

options(repr.plot.width=4.5, repr.plot.height=4)


# theme_set(theme_minimal())

data <- data.frame(
  years  = c(2, 5, 6, 9, 10,10,11,11,12), 
  salary = c(24,30,24,29,31,24,21,41,60)
)
```

```r
data %>% 
  ggplot(aes(x = years, y = salary)) + 
  geom_point() + 
  labs(
    title = 'Earnings Increase with Years of Education', 
    x = 'Years of Employment', 
    y = 'Salary'
  )
ggsave('salary_no_summary.pdf', device = 'pdf')
```

```r
data %>% 
  ggplot(aes(x = years, y = salary)) + 
  geom_point() + 
  stat_smooth(method = 'lm', se = F, color = 'red') + 
  labs(
    title = 'Earnings Increase with Years of Education', 
    x = 'Years of Employment', 
    y = 'Salary'
  )
ggsave("salary_linear.pdf", device = 'pdf')
```

```r
data %>% 
  ggplot(aes(x = years, y = salary)) + 
  geom_point() + 
  stat_smooth(method = 'lm', formula = y ~ exp(x), se = F, color = 'red') + 
  labs(
    title = 'Earnings Increase with Years of Education', 
    x = 'Years of Employment', 
    y = 'Salary'
  )  
ggsave("salary_exponential.pdf", device = 'pdf')
```

```r
data %>% 
  ggplot(aes(x = years, y = salary)) + 
  geom_point() + 
  stat_smooth(method = 'lm', formula = y ~ x + I(x==12), se = F, color = 'red') + 
  labs(
    title = 'Earnings Increase with Years of Education', 
    x = 'Years of Employment', 
    y = 'Salary'
  )  

ggsave("salary_grad.pdf", device = 'pdf')
```
